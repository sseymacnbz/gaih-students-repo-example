```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=10000, n_features=8,
                          n_classes=2, n_informative=5,
                          class_sep=2, random_state=42)
```

```python
type(X)
```

```python
col = ["f1","f2","f3","f4","f5","f6","f7","f8"]
df = pd.DataFrame(X, columns = col)
```

```python
df
```

```python
target = pd.DataFrame(y, columns = ["class"])
```

```python
df = df.join(target)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
plt.figure(figsize=(8,7))
for i, column in enumerate(df.columns, 1):
    plt.subplot(3,3,i)
    sns.distplot(df[column])
    
plt.tight_layout()
```

```python
sns.heatmap(df.corr())
```

```python
sns.boxplot(data=df)
```

```python
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
iqr = q3 - q1
```

```python
iqr
```

```python
outliers = list(set(np.where((df < (q1 - 1.5*iqr)) | (df > (q3 + 1.5*iqr)))[0]))
#newDf=df[~(df < (q1 - 1.5*iqr)) | (df > (q3 + 1.5*iqr))]
```

```python
len(outliers)
```

```python
newDf = df.drop(outliers, axis = 0).reset_index(drop = True)
```

```python
sns.boxplot(data=newDf)
```

```python
X, y = newDf.iloc[:,:-1], newDf.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, 
                                                    random_state=123)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth=5, criterion="gini",random_state=42)
clf.fit(X_train,y_train)
```

```python
from sklearn.externals.six import StringIO
from IPython.display import Image
import pydotplus
from sklearn.tree import export_graphviz
feature_cols = ['f1','f2','f3','f4','f5','f6','f7','f8']

dot_data=StringIO()
export_graphviz(clf, out_file=dot_data,
               filled=True, rounded=True,
               special_characters=True, feature_names=X.columns,
               class_names=['0', '1'])

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png('Output.png')
Image(graph.create_png())
```

```python
plt.figure(figsize=(12,8))
importance = clf.feature_importances_
sns.barplot(x=importance, y=X.columns)
```

```python
from sklearn.metrics import precision_score, recall_score, accuracy_score, classification_report, f1_score, confusion_matrix
pred = clf.predict(X_test)
cm = confusion_matrix(y_test,pred)
ax = sns.heatmap(cm, square=True, annot=True, cbar=False)
ax.xaxis.set_ticklabels(['1','0'],fontsize = 12, rotation=0)
ax.yaxis.set_ticklabels(['1','0'],fontsize = 12, rotation=0)
ax.set_xlabel('Predicted Labels',fontsize = 15)
ax.set_ylabel('True Labels',fontsize = 15)
print(classification_report(y_test,pred))
```
